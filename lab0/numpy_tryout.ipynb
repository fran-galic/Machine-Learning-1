### Bollean Indexing

```python
names = np.array(["Bob", "Joe", "Will", "Bob", "Will", "Joe", "Joe"])
names == "Bob"
```

```python
print(names[names == "Bob"]) # unutra se zapravo vrait polje
boolean_mask = names == "Joe"
names[boolean_mask]
```

```python
# di to moze biti super korisno
names = np.array(["Bob", "Joe", "Will", "Bob", "Will", "Joe", "Joe"])
data_height = np.array([180, 120, 145, 123, 456, 567, 789])
data_height[names == "Bob"]                     # visine svih Bobova
```

```python
# takoder se nakon filtriranja odmah moze i indeksiran
data_height[names == "Bob"][0]
```

```python
data_height[names != "Bob"]
```

```python
print(~(names == "Bob"))
print((names == "Bob") | (names == "Will"))
print((names == "Bob") & (names == "Will"))
```

```python
# Indeksiranje s boolean irazom, vrijednost s desnse strane ce se dodijeliti svima koji s lijeve strane imaju True
names = names.astype("<U10")
names[names == "Bomb"] = "Bombastic"
names
```

#### Fancy indexing

```python
# kada se preda lista indexa koji specificiraju po kojem tocno redu zelimo da se vrait elementi unutar ndarraya
a = np.array([[i for j in range(5)] for i in range(5)])
b = np.array([[j for j in range(5)] for i in range(5)])
print(a)
print(a[[2, 4, 0]])   # opet cemo dobiti ndarray ali kopiju
print(b)
print(b[:, [2, 1, 0]])  # moze raidit i z aredke i za stupce, tj za sve dimenzije
```

### Operacije s matricama, transponiranje , reshaping i to

```python
a = np.arange(16).reshape((4, 4))
a
```

```python
b = a.T         # Transponiranje je i dalje view na orginal
print(b)
b[0][1] = 44
print(b)
print(a)
```

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([1, 2, 3, 4, 5])
print(a.shape)
print(b.T.shape)
print(a.T @ b)
print(a @ b)
```

```python
a = np.array([1, 2, 3, 4, 5]).reshape((1, 5))
b = np.array([1, 2, 3, 4, 5]).reshape((1, 5))
print(a)
print(b.T)
print(a @ b.T)
print(a.T @ b)
print(np.dot(a.T, b)) # ista stvar
```

### Univerzalne funkcije

Univerzalne funkcije ili ti *ufunc* su samo funkcije koje djeluju na cijeli ndarray i vracaju ndarray

Ufunc-s u pravilu uvijek vracaju Novo polje, tj ne vracaju view nego novo polje

```python
a = [1, 2, 3, 4, 5]
np.sqrt(a) # ista stavr samo vrait ndarray sa korijenovanvim vrijednostima
np.exp(a)
```

```python
a = np.array([1, 2.3, 3, 4, 5, 6])
b = np.array([5, 5, 5, 5, 5, 5])
np.maximum(a, b)            # binary ufunc,   ono gore su bili unary ufuncs
np.minimum(a, b)
print(a + b)
```

### neke ostale stavri

```python
# efektinvo if else condition
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
rezz = np.where(cond, xarr, yarr) # uzmi x tamo di je uvjet istinit inace y
print(rezz)
```
