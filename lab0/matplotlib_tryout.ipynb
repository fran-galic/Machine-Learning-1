```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
from sklearn.linear_model import LogisticRegression, Ridge
from sklearn.svm import SVC

# 1. DEFINICIJA SKUPA PODATAKA
# Ulazni vektori X = (x1, x2)
X = np.array([
    [-2, 3], [-1, 2], [0, 1], [0, 0], [1, 1],  # Negativna klasa
    [1, -1], [2, 0]                           # Pozitivna klasa
])

# Oznake za SVM/LinReg (y' = {-1, 1})
y_neg_pos = np.array([-1, -1, -1, -1, -1, 1, 1])

# Oznake za LogReg (y = {0, 1})
y_zero_one = np.array([0, 0, 0, 0, 0, 1, 1])

# Definiramo boje za plot (plava za negativnu, narančasta za pozitivnu)
cmap_points = ListedColormap(['#1f77b4', '#ff7f0e']) # Plava i Narančasta
cmap_light = ListedColormap(['#b4c0e6', '#ffe0b2'])  # Svijetlo plava i Svijetlo narančasta

# 2. INICIJALIZACIJA I TRENIRANJE MODELA

# a) SVM (Hard Margin, Linear Kernel) - Koristi C=1e10 za tvrdu marginu
svm = SVC(kernel='linear', C=1e10, random_state=42)
svm.fit(X, y_neg_pos)

# b) Logistička Regresija (Konvergirana)
log_reg = LogisticRegression(solver='liblinear', max_iter=1000, tol=1e-6, random_state=42)
log_reg.fit(X, y_zero_one)

# c) Linearna Regresija
lin_reg = Ridge(alpha=1e-6)
lin_reg.fit(X, y_neg_pos)


# 3. FUNKCIJA ZA CRTANJE GRANICA
# Kreiranje mreže (grid) točaka za ispunu boja i granica
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                     np.arange(y_min, y_max, 0.05))

def plot_boundary(ax, model, X, y_data, title, is_lin_reg=False):
    # --- Korak 1: Predviđanje na Gridu (za pozadinsku boju) ---
    if is_lin_reg:
        # LinReg: Klasifikacijsko pravilo je y_pred >= 0 => Klasa 1
        Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = np.where(Z >= 0, 1, 0)
    elif isinstance(model, SVC):
        # SVM: Oznake -1 i 1 pretvaramo u 0 i 1 za boju
        Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = np.where(Z == -1, 0, 1)
    else: # LogReg: Oznake 0 i 1
        Z = model.predict(np.c_[xx.ravel(), yy.ravel()])

    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.8)

    # --- Korak 2: Crtanje Linije Odluke (Granice) ---
    # Granica Odluke: w1*x1 + w2*x2 + b = 0
    if not is_lin_reg:
        w1, w2 = model.coef_[0]
        b = model.intercept_[0]
    else: # LinReg
        w1, w2 = model.coef_
        b = model.intercept_

    # Izračun y-koordinate linije za zadane x-koordinate
    # x2 = (-w1/w2)*x1 - (b/w2)
    x1_line = np.array([x_min, x_max])
    x2_line = (-w1 / w2) * x1_line - (b / w2)

    ax.plot(x1_line, x2_line, 'k-', linewidth=2, label='Granica Odluke')

    # --- Korak 3: Iscrtavanje Izvornih Točaka ---
    # Koristimo y_zero_one za sva plotiranja jer to najbolje odgovara cmap_points
    ax.scatter(X[:, 0], X[:, 1], c=y_zero_one, cmap=cmap_points,
               edgecolors='k', s=80, label='Podaci (0=Plavo, 1=Narančasto)')

    ax.set_title(title, fontsize=12) # Smanjen font za bolji prikaz
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.grid(True)


# 4. KREIRANJE GRAFIKONA ZA TRI MODELA
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Granice Odluke za Linearno Separabilan Skup Točaka', fontsize=16)

# a) SVM
plot_boundary(axes[0], svm, X, y_neg_pos,
              f'1. Linear SVM (Hard Margin)\nGranica: {svm.coef_[0][0]:.2f}x1 + {svm.coef_[0][1]:.2f}x2 + {svm.intercept_[0]:.2f} = 0')

# b) Logistička Regresija
plot_boundary(axes[1], log_reg, X, y_zero_one,
              f'2. Logistička Regresija (Konverg.)\nGranica: {log_reg.coef_[0][0]:.2f}x1 + {log_reg.coef_[0][1]:.2f}x2 + {log_reg.intercept_[0]:.2f} = 0')

# c) Linearna Regresija (ISPRAVLJENA LINIJA ZA NASLOV)
plot_boundary(axes[2], lin_reg, X, y_zero_one,
              f'3. Linearna Regresija (y_pred >= 0)\nGranica: {lin_reg.coef_[0]:.2f}x1 + {lin_reg.coef_[1]:.2f}x2 + {lin_reg.intercept_:.2f} = 0',
              is_lin_reg=True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression, Ridge
from sklearn.svm import SVC

# Podaci (ponovno definiranje za samostalan rad koda, iako su isti kao prije)
X = np.array([[-2, 3], [-1, 2], [0, 1], [0, 0], [1, 1], [1, -1], [2, 0]])
y_neg_pos = np.array([-1, -1, -1, -1, -1, 1, 1])
y_zero_one = np.array([0, 0, 0, 0, 0, 1, 1])

# Točka za testiranje
P0 = np.array([2, 0])
x0, y0 = P0

# 1. Treniranje Modela (s istim parametrima)
svm = SVC(kernel='linear', C=1e10, random_state=42).fit(X, y_neg_pos)
log_reg = LogisticRegression(solver='liblinear', max_iter=1000, tol=1e-6, random_state=42).fit(X, y_zero_one)
lin_reg = Ridge(alpha=1e-6).fit(X, y_neg_pos)

# Funkcija za izračun udaljenosti
def calculate_distance(model_name, w, b, x, y):
    # Geometrijska udaljenost
    numerator = np.abs(w[0] * x + w[1] * y + b)
    denominator = np.sqrt(w[0]**2 + w[1]**2)
    distance = numerator / denominator

    print(f"--- {model_name} ---")
    print(f"Koeficijenti (w1, w2, b): ({w[0]:.4f}, {w[1]:.4f}, {b[0]:.4f})")
    print(f"Udaljenost točke (2, 0) do granice: {distance[0]:.4f}")

    if model_name == "Linear SVM":
        # Funkcionalna margina SVM-a je w*x + b
        func_margin = w[0] * x + w[1] * y + b
        print(f"Funkcionalna margina (w*x+b): {func_margin[0]:.4f} (Treba biti >= 1)")
    print("-" * 25)

# 2. Izračun udaljenosti za svaki model
print("\n*** Udaljenost točke (2, 0) do granice odluke ***")

# a) SVM
calculate_distance("Linear SVM", svm.coef_[0], svm.intercept_, x0, y0)

# b) Logistička Regresija
calculate_distance("Logistička Regresija", log_reg.coef_[0], log_reg.intercept_, x0, y0)

# c) Linearna Regresija
calculate_distance("Linearna Regresija", lin_reg.coef_, lin_reg.intercept_[np.newaxis], x0, y0)
```
