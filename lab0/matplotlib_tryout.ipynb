```python
a = np.random.standard_normal(200)
b = np.random.standard_normal(200)
plt.scatter(a, b)          # scatter je tockasti graf
plt.show()
```

```python
plt.boxplot(a)
plt.title("Neki naslov")
plt.grid(False)
plt.xlabel("neki label za x os")
plt.ylabel("neki label za y os")
plt.legend(["linija 1"])
plt.show()
```

```python
ls = np.linspace(0, 100, 100)
def f(x):
    return x**2
ls = f(ls)
print(ls)
```

```python
a = np.linspace(1, 100, 100)
print(a)
plt.plot(a, f(a), "+")
```

```python
plt.plot(a, f(a), "r+" , a, -f(a) + 1000, "b--")
plt.legend("bg", loc="center right")
```

```python
a = np.linspace(1, 100, 100)
plt.plot(a, f(a), label="f(a)", color="red")
plt.plot(a, 1 - f(a), label="1 - f(a)", color="blue")
plt.legend(loc="best")
plt.xlim([40, 60])
plt.show()
```
